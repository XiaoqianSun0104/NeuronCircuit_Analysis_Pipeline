### Define some functions

```python
# get rid of incorrect neuron and non-target neuron for trace file
#----------------------------------------------------------------------------------
def process_(trace_path):
    
    #read csv
    df = pd.read_csv(trace_path)
    
    #profecc column in df file
    column_list = df.columns.tolist()
    column_list = [column.strip() for column in column_list]
    column_list[0] = 'Frame'
    df.columns = column_list
    
    # drop the first row, cell-undecide
    df =df.drop([0])

    #convert object dtype to float
    df = df.astype('float')
    
    #set Frame as index
    df = df.set_index('Frame')
    return(df)



# process frame list
# event file always start from 10.1s, but some trace frame start from 0, some from 741.19
#----------------------------------------------------------------------------------

def process_frame(df):
    # get frame list
    frame_list = df.index.values.tolist()
    
    # get starting value
    a = frame_list[0]
    new_frame_list = [(i-a) for i in frame_list]
    
    return(new_frame_list)


# process event_label list
# event+label starts from 10, but trace start from 0, deduct 10 from event_start&end time
#----------------------------------------------------------------------------------
def process_event_interval(df):
    new_col = []
    for col in df.columns.tolist():
        new_col.append(col.strip())
    df.columns = new_col
    
    event_list = df[['From Second','To Second']].values.tolist()
    startTime = df['From Second'].values.tolist()[0]
    new_event_list = []
    for event in event_list:
        new_event = [x -startTime for x in event]
        new_event_list.append(new_event)
    return(new_event_list)


#function to sepatate sessions of one sample
#----------------------------------------------------------------------------------

#use separate_session(df)[0/1/2] to get each session
def separate_session(df):
    
    check_point = []

    frame_list = df.index.values.tolist()

    for i in range(len(df)):
        if i < len(df)-1:
            if (frame_list[i+1] - frame_list[i]) > 3:
                check_point.append(frame_list[i+1])

    df_list = []
    
    if len(check_point) == 2:
        # separate
        df_1 = df[df.index < check_point[0]]
        df_2 = df[(df.index < check_point[1]) & (df.index >= check_point[0])]
        df_3 = df[df.index >= check_point[1]]
        df_list.extend([df_1, df_2, df_3])
    elif len(check_point) == 3:
        # separate
        df_1 = df[df.index < check_point[0]]
        df_2 = df[(df.index < check_point[1]) & (df.index >= check_point[0])]
        df_3 = df[(df.index < check_point[2]) & (df.index >= check_point[1])]
        df_4 = df[df.index >= check_point[2]]
        df_list.extend([df_1, df_2, df_3, df_4])
    
    return(df_list)
```

```python
# get enevt label for an event file based on different mice and session
# for each session, we have separate event file
#----------------------------------------------------------------------------------
def get_event_label(event_df, sniff_side):
    
    # get event status list
    Event_status = event_df['Event'].values.tolist()
    # compact event label list for three sessions
    event_label = []

    if sniff_side == "right":
        for i in Event_status:   
            if i.strip() == 'Area:Mouse 1 Center In Box':
                event_label.append(10)
            elif i.strip() == 'Area:Mouse 1 Center In Left' :
                event_label.append(13)
            elif i.strip() == 'Area:Mouse 1 Center In Middle':
                event_label.append(12)
            elif i.strip() == 'Area:Mouse 1 Center In Right': # social zone
                event_label.append(11)
            elif i.strip() == 'Area:Mouse 1 Nose In s1':
                event_label.append(21)
            elif i.strip() == 'Area:Mouse 1 Nose In s2':
                event_label.append(22)
            elif i.strip() == 'Mouse 1 Approaching Area s1':
                event_label.append(31)
            elif i.strip() ==  'Mouse 1 Approaching Area s2':
                event_label.append(32)
            elif i.strip() == "Mouse 1 Grooming":
                event_label.append(41)
            elif i.strip() == "Mouse 1 sniffing On s1":
                event_label.append(51)
            elif i.strip() == "Mouse 1 sniffing On s2":
                event_label.append(52)
            else:
                event_label.append(90)
    else :   #(left + no)
        for i in Event_status:   
            if i.strip() == 'Area:Mouse 1 Center In Box':
                event_label.append(10)
            elif i.strip() == 'Area:Mouse 1 Center In Left' : # social zone (s1 + no)
                event_label.append(11)
            elif i.strip() == 'Area:Mouse 1 Center In Middle':
                event_label.append(12)
            elif i.strip() == 'Area:Mouse 1 Center In Right': # object zone
                event_label.append(13)
            elif i.strip() == 'Area:Mouse 1 Nose In s1':
                event_label.append(21)
            elif i.strip() == 'Area:Mouse 1 Nose In s2':
                event_label.append(22)
            elif i.strip() == 'Mouse 1 Approaching Area s1':
                event_label.append(31)
            elif i.strip() ==  'Mouse 1 Approaching Area s2':
                event_label.append(32)
            elif i.strip() == "Mouse 1 Grooming":
                event_label.append(41)
            elif i.strip() == "Mouse 1 sniffing On s1":
                event_label.append(51)
            elif i.strip() == "Mouse 1 sniffing On s2":
                event_label.append(52)
            else:
                event_label.append(90)
        
                
    return(event_label)



# get sniff label for an sniff file based on different mice and session
# for each session, we have separate sniff file
#----------------------------------------------------------------------------------
def get_sniff_label(sniff_df, sniff_side):
    
    # get sniff status list
    sniff_status = sniff_df['Event'].values.tolist()
    # compact sniff label list for three sessions
    sniff_label = []
    
    if sniff_side == "right":
        for i in sniff_status:   
            if i.strip() == 'Mouse 1 sniffing On s1':
                sniff_label.append(61)
            elif i.strip() == 'Mouse 1 sniffing On s2' :
                sniff_label.append(60) # new object side
            elif i.strip() == 'No sniffing' :
                sniff_label.append(62)
            else :
                sniff_label.append(90)
    else : #(left + no)
        for i in sniff_status:   
            if i.strip() == 'Mouse 1 sniffing On s1':
                sniff_label.append(60) # new object side
            elif i.strip() == 'Mouse 1 sniffing On s2' :
                sniff_label.append(61) 
            elif i.strip() == 'No sniffing' :
                sniff_label.append(62)
            else :
                sniff_label.append(90)
        
    return(sniff_label)



# process events file, add event label for string label
#----------------------------------------------------------------------------------
def add_event_label(trace_df, event_path, sniff_path, sniff_side):

    frame_list = process_frame(trace_df)

    #get event dataframe, get event intervals, get event label
    event_df = pd.read_csv(event_path)
    interval_list = process_event_interval(event_df)
    event_label = get_event_label(event_df, sniff_side)
    
    # get sniff df, get sniff intervals, get sniff label
    sniff_df = pd.read_csv(sniff_path)
    sniff_interval_list = process_event_interval(sniff_df)
    sniff_label = get_sniff_label(sniff_df, sniff_side)

    frame_label_list = []
    for frame in frame_list:
        frame_label = str()
        if frame > 5:
            # begin to define frame label
            for i in range(len(interval_list)):
                if min(interval_list[i])  <= frame <= max(interval_list[i]) :
                    frame_label = frame_label+ str(event_label[i])
            
            for j in range(len(sniff_interval_list)):
                if min(sniff_interval_list[j])  <= frame <= max(sniff_interval_list[j]) :
                    frame_label = frame_label+ str(sniff_label[j])
                

        # convert str to int
        if len(frame_label) == 0:
            frame_label = np.nan
        else:
            frame_label = int(frame_label)

        # append frame label to list
        frame_label_list.append(frame_label)
            
    # add label list to trace df 
    new_df = trace_df.copy()
    new_df['Frame_Label']  =  frame_label_list
    new_df = new_df.dropna()
    
    #print('The df shape of trace df with label:',new_df.shape)
    return(new_df)
```

### Call

```python
for i in range(len(WT_list)):
    mouse_summary = WT_list[i]
    
    mouse_name = mouse_summary.split('_')[0]
    mouse_type = mouse_summary.split('_')[1]
    
    mouse = WT_mouse[i]
    print(mouse_name)
    
    session_list = Social_tip[Social_tip['ID'] == mouse_name]['TRAIL'].values.tolist()
    sniff_list = Social_tip[Social_tip['ID'] == mouse_name]['Social side'].values.tolist()

    # process trace
    trace = os.path.join(root_path, mouse, "trace.csv")
    df = process_(trace)
    df_sessions = separate_session(df)
    
    
    # get folder path for each mouse
    mouse_folder = os.path.join(root_path, mouse)
    neuron_list = []
    for filename in os.listdir(mouse_folder):
        if filename.endswith(".tiff") and filename.startswith("image_C") : 
            neuron_list.append(filename.split('.')[0].split('_')[1])    
    
    
    # add event label to each session
    for i in range(len(df_sessions)):
        df_session = df_sessions[i]

        event_name = "event_" + session_list[i] +"_" + sniff_list[i] + ".csv"
        event_session_path = os.path.join(root_path, mouse, event_name)

        sniff_name = "sniff_" + session_list[i] +"_" + sniff_list[i] + ".csv"
        sniff_session_path = os.path.join(root_path, mouse, sniff_name)

        df_session_frame_label = add_event_label(df_session, event_session_path, sniff_session_path, sniff_list[i])
        df_keep = df_session_frame_label[neuron_list+["Frame_Label"]]
        df_session_name = "trace_" + session_list[i] + "_" + sniff_list[i] + ".csv"
        df_keep.to_csv(os.path.join(root_path, mouse, df_session_name))
        
```
